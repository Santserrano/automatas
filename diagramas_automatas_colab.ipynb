# Diagramas de Automatas en Colab

Este notebook genera tablas de transicion y diagramas para:
1. AFD (determinista)
2. AFND (no determinista)
3. Automata de pila (PDA)

```python
import pandas as pd
from graphviz import Digraph
from IPython.display import display

def mostrar_tabla(df, titulo):
    print(f"\n{titulo}")
    display(df)

def mostrar_diagrama(dot):
    display(dot)
```

## 1) AFD - Reconoce exactamente la cadena: hola mundo

```python
# Tabla de transicion AFD
afd_rows = [
    ["q0", "h", "q1"],
    ["q1", "o", "q2"],
    ["q2", "l", "q3"],
    ["q3", "a", "q4"],
    ["q4", "espacio", "q5"],
    ["q5", "m", "q6"],
    ["q6", "u", "q7"],
    ["q7", "n", "q8"],
    ["q8", "d", "q9"],
    ["q9", "o", "q10"],
    ["q10", "otro", "qT"],
    ["qT", "cualquiera", "qT"]
]
afd_df = pd.DataFrame(afd_rows, columns=["Estado", "Simbolo", "Siguiente"])
mostrar_tabla(afd_df, "Tabla de transicion AFD")

# Diagrama AFD
afd = Digraph("AFD", format="png")
afd.attr(rankdir="LR")

for s in ["q0","q1","q2","q3","q4","q5","q6","q7","q8","q9","q10","qT"]:
    if s == "q10":
        afd.node(s, shape="doublecircle")
    else:
        afd.node(s, shape="circle")

afd.node("inicio_afd", shape="point")
afd.edge("inicio_afd", "q0")

afd.edge("q0", "q1", label="h")
afd.edge("q1", "q2", label="o")
afd.edge("q2", "q3", label="l")
afd.edge("q3", "q4", label="a")
afd.edge("q4", "q5", label="espacio")
afd.edge("q5", "q6", label="m")
afd.edge("q6", "q7", label="u")
afd.edge("q7", "q8", label="n")
afd.edge("q8", "q9", label="d")
afd.edge("q9", "q10", label="o")
afd.edge("q10", "qT", label="otro")
afd.edge("qT", "qT", label="cualquiera")

mostrar_diagrama(afd)
```

## 2) AFND - Reconoce hola o mundo con una bifurcacion epsilon

```python
# Tabla de transicion AFND
afnd_rows = [
    ["p0", "eps", "{p1, p6}"],
    ["p1", "h", "{p2}"],
    ["p2", "o", "{p3}"],
    ["p3", "l", "{p4}"],
    ["p4", "a", "{pf}"],
    ["p6", "m", "{p7}"],
    ["p7", "u", "{p8}"],
    ["p8", "n", "{p9}"],
    ["p9", "d", "{p10}"],
    ["p10", "o", "{pf}"]
]
afnd_df = pd.DataFrame(afnd_rows, columns=["Estado", "Simbolo", "Delta"])
mostrar_tabla(afnd_df, "Tabla de transicion AFND")

# Diagrama AFND
afnd = Digraph("AFND", format="png")
afnd.attr(rankdir="LR")

for s in ["p0","p1","p2","p3","p4","p6","p7","p8","p9","p10","pf"]:
    if s == "pf":
        afnd.node(s, shape="doublecircle")
    else:
        afnd.node(s, shape="circle")

afnd.node("inicio_afnd", shape="point")
afnd.edge("inicio_afnd", "p0")

afnd.edge("p0", "p1", label="eps")
afnd.edge("p0", "p6", label="eps")
afnd.edge("p1", "p2", label="h")
afnd.edge("p2", "p3", label="o")
afnd.edge("p3", "p4", label="l")
afnd.edge("p4", "pf", label="a")
afnd.edge("p6", "p7", label="m")
afnd.edge("p7", "p8", label="u")
afnd.edge("p8", "p9", label="n")
afnd.edge("p9", "p10", label="d")
afnd.edge("p10", "pf", label="o")

mostrar_diagrama(afnd)
```

## 3) Automata de pila (PDA) - Lenguaje a^n b^n, n >= 1

```python
# Tabla de transicion PDA
pda_rows = [
    ["q0", "a", "Z", "q0", "Z,A"],
    ["q0", "a", "A", "q0", "A,A"],
    ["q0", "b", "A", "q1", "eps"],
    ["q1", "b", "A", "q1", "eps"],
    ["q1", "eps", "Z", "q2", "Z"]
]
pda_df = pd.DataFrame(pda_rows, columns=["Estado", "Entrada", "Tope", "Siguiente", "Push"])
mostrar_tabla(pda_df, "Tabla de transicion PDA")

# Diagrama PDA
pda = Digraph("PDA", format="png")
pda.attr(rankdir="LR")

for s in ["q0", "q1", "q2"]:
    if s == "q2":
        pda.node(s, shape="doublecircle")
    else:
        pda.node(s, shape="circle")

pda.node("inicio_pda", shape="point")
pda.edge("inicio_pda", "q0")

pda.edge("q0", "q0", label="a, Z / ZA\na, A / AA")
pda.edge("q0", "q1", label="b, A / eps")
pda.edge("q1", "q1", label="b, A / eps")
pda.edge("q1", "q2", label="eps, Z / Z")

mostrar_diagrama(pda)
```
